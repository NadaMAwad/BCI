# Import Modules

```python
import numpy as np
from scipy.io import loadmat  # this is the SciPy module that loads mat-files
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
```

#  Load & Prepare Data

```python
mat = loadmat('Subject1_2D-Copy1.mat')  # load mat-file
mat #Show File Content
```

```python
# File is Dictionary so Now we know its content
mat.keys()
```

```python
# each key refer to list of arrays 
# so we extract it from dict and convert it to 2d array called (data farme in pandas)
# and join them together
```

```python
# Data of Left:

Leftf1 =list(mat['LeftForward1'])
Leftf2 =list(mat['LeftForward2'])
Leftf3 =list(mat['LeftForward3'])

Lf1 = pd.DataFrame(list(map(np.ravel, Leftf1)))
Lf2 = pd.DataFrame(list(map(np.ravel, Leftf2)))
Lf3 = pd.DataFrame(list(map(np.ravel, Leftf3)))

DL = pd.concat([Lf1, Lf2, Lf3]) #join them
```

```python
# Data of Right Back & Forwad: 

Rightf1 =list(mat['RightBackward1'])
Rightf2 =list(mat['RightBackward2'])
Rightf3 =list(mat['RightBackward3'])

Rf1 = pd.DataFrame(list(map(np.ravel, Rightf1)))
Rf2 = pd.DataFrame(list(map(np.ravel, Rightf2)))
Rf3 = pd.DataFrame(list(map(np.ravel, Rightf3)))

DR = pd.concat([Rf1, Rf2, Rf3]) #join them
```

```python
# We Use Classification so befor creating model we labled data
```

```python
DL.insert(column = 'lable' , loc= 19,  value ="Left", allow_duplicates = True)
DR.insert(column = 'lable' , loc= 19,  value ="Right", allow_duplicates = True)
```

```python
# join all Data
```

```python
Data = pd.concat([DL,DR])
```

```python
Data
```

```python
# raname columns by brain sites
Data.columns= ['FP1','FP2', 'F3', 'F4', 'C3', 'C4', 'P3', 'P4', 'O1', 'O2', 'F7', 'F8', 'T3', 'T4', 'T5', 'T6', 'FZ', 'CZ', 'PZ',
'lable']
```

```python
Data
```

# Build SVM Model

```python
# Create Model 
svc = SVC()
```

```python
# Select coulmns of futures and lables
X = Data.iloc[: , :19]
Y = Data.lable
```

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

```python
# Fitting the classifier into the Training set

classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
# Predicting the test set results

Y_Pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_Pred)
```
